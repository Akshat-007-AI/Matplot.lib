```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("pokemon.csv")
df
```

```python
df.info()
```

```python
x = df["Speed"]
y = df["Generation"]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(df["Generation"],df["Speed"].sort_values(),
color="orange",
linestyle="--",
linewidth=2,
marker="x",
markersize=8,
label="Generation",
alpha=0.8)
plt.title("Generation vs Speed")
plt.xlabel("Generation")
plt.ylabel("Speed")

plt.show()
```

```python
plt.scatter(df["Speed"],df["Attack"],
c="green",
s=100,
marker="o",
alpha=0.7,
edgecolors="blue",)
plt.xlabel("Speed")
plt.ylabel("Attack")
plt.title("Speed vs Attack")

plt.show()
```

```python
plt.bar(df["Type 1"],df["Speed"],
color="blue",
edgecolor="red",
alpha=0.8)

plt.show()
```

```python
plt.barh(df["Type 1"],df["Speed"],
color="blue",
edgecolor="red",
alpha=0.8)

plt.show()

```

```python
plt.hist(df["Generation"],
bins=30,
color="ORANGE",
edgecolor="BLUE",
density=True,
)

plt.show()
```

```python
plt.boxplot(df["Speed"],
showmeans=True,
patch_artist=True,
)

plt.show()
```

```python
plt.pie(df["Type 1"].value_counts(),
labels=df["Type 1"].value_counts().index,
autopct="%1.1f%%",
startangle=90)

plt.show()
```

```python
m = [1,2,3,4,5]
n = [10,20,15,25,30]

plt.stem(m,n)
plt.title("Stem Plot")
plt.xlabel("m values")
plt.ylabel("n values")
```

```python
g = [1,2,3,4,5]
h = [10,20,30,40,50]
plt.step(g,h)
plt.title("Step Plot")

plt.show()
```

```python
c = [1,2,3,4,5]
d = [10,15,25,20,30]
plt.fill_between(c,d)
plt.title("Area Plot")
plt.show()
```
